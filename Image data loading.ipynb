```python
import tensorflow as tf

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  label_mode=None,
  image_size=(640, 480),
  batch_size=None)

len(train_ds)
```

```python
import tensorflow as tf

def normalize_image(image):
    # Scale the pixel values to the range [0, 1]
    image = tf.cast(image, tf.float32) / 255.0
    return image

# Assuming you have created the train_ds as shown in your code
# You can normalize the images using the map() function
normalized_train_ds = train_ds.map(normalize_image)

# To check the number of batches in the normalized dataset
num_batches = len(normalized_train_ds)
print("Number of batches in the normalized dataset:", num_batches)
```
